#### Створіть свій датасет на основі випадкових даних за допомогою numpy.linspace, та проведіть на ньому лінійну регресію, використовуючи інструменти tensorflow

```python
# імпорт бібліотек
import numpy as np
import tensorflow as tf
import tensorflow.compat.v1 as tfc
import matplotlib.pyplot as plt
```

```python
# генерація випадкових лінійних даних
x = np.linspace(0, 50, 50)
y = np.linspace(0, 50, 50)
```

```python
# додавання шуму до випадкових лінійних даних
x += np.random.uniform(-4, 4, 50)
y += np.random.uniform(-4, 4, 50)
```

```python
# візуалізація випадкових лінійних даних
plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Training Data')
plt.show()
```

```python
# створення і резервація плейсхолдерів
tf.compat.v1.disable_eager_execution()
X = tfc.placeholder('float')
Y = tfc.placeholder('float')
```

```python
# стаорення змінних коефіцієнтів лінійної моделі
W = tf.Variable(np.random.randn(), name = "W")
b = tf.Variable(np.random.randn(), name = "b")
```

```python
# задання гіпотези
y_pred = tf.add(tf.multiply(X, W), b)
```

```python
# задання функції втрат
cost = tf.reduce_sum(tf.pow(y_pred-Y, 2)) / (2 * len(x))
```

```python
learning_rate = 0.01
training_epoch = 1000
```

```python
# задання оптимізатора
optimizer_gradient = tfc.train.GradientDescentOptimizer(learning_rate).minimize(cost)
init = tfc.global_variables_initializer()
```

```python
# навчання моделі
with tfc.Session() as sess:
    sess.run(init)
    for epoch in range(training_epoch):
        for(_x, _y) in zip(x, y):
            sess.run(optimizer_gradient, feed_dict = {X : _x, Y : _y})
            if (epoch + 1) % 50 == 0:
                c = sess.run(cost, feed_dict = {X : x, Y : y})
                print('Epoch', (epoch + 1), ": cost =", c, "W =", sess.run(W), "b =", sess.run(b))
                
    training_cost = sess.run(cost, feed_dict = {X: x, Y: y})
    weight = sess.run(W)
    bias = sess.run(b)
```

```python
# результат роботи моделі
predictions = weight * x + bias
print("Training cost = ",
     training_cost,
     "Weight = ", weight,
     "bias =", bias, "\n")
```

```python
# візуалізація роботи моделі
plt.plot(x, y, "ro", label = "Original data")
plt.plot(x, predictions, label = "Fitted line")
plt.title("Linear Regression Result")
plt.legend()
plt.show()
```
